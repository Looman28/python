```python
import seaborn as sns
import matplotlib.pyplot as plt
iris_data=sns.load_dataset('iris')
iris_data
```

```python
sns.boxplot(x=iris_data['sepal_width'],color='red')
plt.show()
```

```python
sns.boxplot(x=iris_data['sepal_width'],palette=['violet','red','green'],y=iris_data['species'])
plt.show()
```

```python
df=sns.load_dataset('taxis')
```

```python
sns.scatterplot(x=df['fare'],y=df['total'],palette=['blue','black'],hue=df['payment'])
plt.show()
```

```python
sns.stripplot(x=df['fare'],y=df['payment'],palette=['orange','black'],hue=df['color'])
```

```python
sns.swarmplot(x=df['fare'],y=df['payment'],palette=['pink','black'],hue=df['color'])
```

```python
sns.lineplot(x=df['total'],y=df['fare'],hue=['payment'],palette=['black','red'])
```
