```python
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\Looman\Downloads\Wholesale customers data.csv")

df.head()
```

```python
df
```

```python
df.drop(['Channel','Region'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(df)
```

```python
x
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
len(set(dendrogram['color_list']))-1
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3)
cluster=model.fit_predict(x)
```

```python
cluster
```

```python
df
```

```python
group_num=pd.DataFrame(cluster,columns=['Group'])
group_num
```

```python
pd.concat([df,group_num],axis=1)
```

```python
cust_group_data=pd.concat([df,group_num],axis=1)
cust_group_data
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x,cluster)
```

```python
cluster.shape
```

#KMeans

```python
x
```

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range (2,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,11),wcss)
plt.title('The elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model = KMeans(n_clusters = 5, random_state = 28)
groups = model.fit_predict(x)
print(groups)
```

```python
groups
```

```python
groups.shape
```

```python
group_num=pd.DataFrame(cluster,columns=['Group'])
group_num
```

```python
cust_kmeans_data=pd.concat([df,group_num],axis=1)
cust_kmeans_data
```

```python
silhouette_score(x,groups)
```
