```python
import matplotlib.pyplot as plt
for col in cars.columns:
  if col != 'HP':  
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='HP', y=col, data=cars)
    plt.title(f'Scatter Plot of HP vs {col}')
    plt.xlabel('HP')
    plt.ylabel(col)
    plt.show()
```
